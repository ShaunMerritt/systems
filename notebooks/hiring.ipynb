```python
from systems.parse import parse
from systems.viz import as_dot
from IPython.core.display import HTML
```

```python
spec = """

[Candidate] > Recruiters(3, 7) @ 1

[Candidate] > Prospect    @ Recruiters * 3
Prospect    > Screen      @ 0.5
Screen      > Onsite      @ 0.5
Onsite      > Offer       @ 0.25
Offer       > Accept      @ 0.5
Accept      > Hired       @ 1.0

"""
model = parse(spec)
```

```python
as_dot(model)
```

```python
results = model.run(rounds=10)
rendered = model.render_html(results)
HTML(rendered)
```

```python
from bokeh.plotting import figure, output_notebook, show
output_notebook()
```

```python
col = "Offer"
x = list(range(len(results)))
y = [row[col] for row in results]
p = figure(title=col)
p.line(x, y)
show(p)
```
